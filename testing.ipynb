```python
import cv2
import torch
from ultralytics import YOLO

# Load the YOLO model
model_path = 'best.pt'  # Replace with your model path
model = YOLO(model_path)

# Load the image
image_path = 'test.jpg'  # Replace with your test image path
image = cv2.imread(image_path)

# Get predictions
results = model(image)

# Define keypoint pairs for skeletal representation
POSE_PAIRS = [
    [0, 1], [1, 2], [2, 3], [3, 4],
    [0, 5], [5, 6], [6, 7], [7, 8],
    [0, 9], [9, 10], [10, 11], [11, 12],
    [0, 13], [13, 14], [14, 15], [15, 16],
    [0, 17], [17, 18], [18, 19], [19, 20]
]

# Draw keypoints and lines
for person_index in range(results[0].keypoints.shape[0]):
    keypoints = results[0].keypoints.xy[person_index]  # Get keypoints for this person

    for point in keypoints:  # Iterate over keypoints
        x, y = point  # Get x, y coordinates
        # Draw the keypoint as a circle
        cv2.circle(image, (int(x), int(y)), 5, (0, 255, 0), -1)  # Green color

    # Draw lines between keypoints
    for pair in POSE_PAIRS:
        pointA = keypoints[pair[0]]
        pointB = keypoints[pair[1]]

        # Draw a line if both points exist
        cv2.line(image, (int(pointA[0]), int(pointA[1])), (int(pointB[0]), int(pointB[1])), (255, 0, 0), 2)  # Red color

# Display the image
cv2.imshow("Keypoints and Skeleton", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

# Load the YOLO model
model_path = 'best.pt'  # Replace with your model path
model = YOLO(model_path)

# Define keypoint pairs for skeletal representation
POSE_PAIRS = [
    [0, 1], [1, 2], [2, 3], [3, 4],
    [0, 5], [5, 6], [6, 7], [7, 8],
    [0, 9], [9, 10], [10, 11], [11, 12],
    [0, 13], [13, 14], [14, 15], [15, 16],
    [0, 17], [17, 18], [18, 19], [19, 20]
]

# Start capturing video from the webcam
cap = cv2.VideoCapture(0)  # 0 is the default camera; change if necessary

while True:
    ret, frame = cap.read()  # Capture frame-by-frame
    if not ret:
        break

    # Get predictions
    results = model(frame)

    # Draw keypoints and lines
    for person_index in range(results[0].keypoints.shape[0]):
        keypoints = results[0].keypoints.xy[person_index]  # Get keypoints for this person

        for point in keypoints:  # Iterate over keypoints
            x, y = point  # Get x, y coordinates
            # Draw the keypoint as a circle
            cv2.circle(frame, (int(x), int(y)), 5, (0, 255, 0), -1)  # Green color

        # Draw lines between keypoints
        for pair in POSE_PAIRS:
            pointA = keypoints[pair[0]]
            pointB = keypoints[pair[1]]

            # Draw a line if both points exist
            cv2.line(frame, (int(pointA[0]), int(pointA[1])), (int(pointB[0]), int(pointB[1])), (255, 0, 0), 2)  # Red color

    # Display the resulting frame
    cv2.imshow("Camera Feed - Keypoints and Skeleton", frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close windows
cap.release()
cv2.destroyAllWindows()
```
